```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
data=[]
for j in tqdm(range(1,11)):
    link="https://quotes.toscrape.com/page/"+str(j)
    res=requests.get(link)
    soup=BeautifulSoup(res.text,"html.parser")
    for i in soup.find_all("div",class_="quote"):
        quote=i.find("span",class_="text").text[1:-1]
        author=i.find("small").text
        author_id=i.find("a").get("href")
        tags=[]
        for i in i.find_all("a",class_="tag"):
            tags.append(i.text)
        tags=",".join(tags)   
        data.append([quote,author,author_id,tags])
d=pd.DataFrame(data,columns=["quote","author","author_id","tags"])
d["Author_link"]="https://quotes.toscrape.com"+d["author_id"]
d.to_csv("Quotes.csv",index=False)
```

```python
paths=[]
d=pd.read_csv("Quotes.csv")
del d["quote"]
del d["author_id"]
del d["tags"]
d=d.drop_duplicates()
#d1= d[d["author"]!="Jorge Luis Borges"]
#d2=d1[d1["author"]!="Haruki Murakami"]
for i in tqdm(d.values):
    name="Authors/" + i[0] + ".txt"
    paths.append(name)
    link=i[1]
    res=requests.get(link)
    soup=BeautifulSoup(res.text,'html.parser')
    t=(soup.find("div",class_="author-details").text)
    f=open(name,'w',encoding="utf-8")
    f.write(t)
    f.close()
d["paths"]=paths
d.to_csv("Authors.csv",index=False)    
```
